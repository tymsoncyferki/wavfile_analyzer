```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from analysis import FileAnalyzer
```

```python
import os

input_folder = "data/music"

df = pd.DataFrame()
i = 0
for filename in os.listdir(input_folder):
    i += 1
    if filename.lower().endswith(".wav"):
        wav_path = os.path.join(input_folder, filename)
        analyzer = FileAnalyzer(wav_path)
        params = analyzer.calculate_parameters()
        param_df = pd.DataFrame(params, index=[i])
        f0, _ = analyzer.f0()
        param_df['F0 Median'] = np.median(f0)
        df = pd.concat([df, param_df])
```

```python
df
```

```python
np.round(df.drop(['Audio type'], axis=1).mean(), 3)
```
